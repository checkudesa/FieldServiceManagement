## Implementar una heurística inicial para el problema.

### Heurística 1°

Una primera heurística consiste en organizar los productos en orden descendente según su relación beneficio/peso y agregarlos a la mochila en ese orden hasta alcanzar la capacidad máxima permitida. Cuando se llega al punto en el que no se puede agregar más, se calcula la capacidad restante disponible y se buscan los productos que sean iguales o menores en peso para llenar la mochila, siempre dando prioridad a los que ofrecen la mejor relación beneficio/peso.

```python
def indice_mochila (items, capacidad):

    # Calculamos el índice de beneficio sobre peso.
    # Guardamos el índice, el beneficio y el peso.
    beneficio_relativo = [(item[0]/item[1], item[0], item[1]) for item in items]

    # Ordenamos de manera descendente
    beneficio_relativo.sort(reverse=True)

    beneficio_total = 0 # Vamos a ir sumando el beneficio acumulado
    mochila = [] # Vamos a ir guardando los ítems que incluimos en la mochila

    for indice, beneficio, peso in beneficio_relativo:

        # Mientras la capacidad residual permita incluir el producto entero
        if capacidad >= peso:
            # Sumamos el beneficio del producto y restamos su peso de la capacidad
            beneficio_total += beneficio
            mochila.append((beneficio, peso))
            capacidad -= peso

        # Si el peso supera la capacidad disponible, continuar con el próximo item
        else:
            continue

    return beneficio_total, mochila, capacidad
```

### Heurística 2°

Una segunda heurística consiste en ordenar en forma descendente los ítems por el índice beneficio/peso, pero comenzar a llenar la mochila con aquellos menos pesados (siempre priorizando los de mayor beneficio relativo) para incluir de esta manera la mayor cantidad de ítems en la mochila.

```python
def peso_mochila (items, capacidad):

    # Calculamos el índice de beneficio sobre peso.
    # Guardamos el índice, el beneficio y el peso.
    beneficio_relativo = [(item[0]/item[1], item[0], item[1]) for item in items]

    # Ordenamos la lista por peso de manera ascendente
    # En caso de igualdad de peso, por índice de beneficio/peso de manera descendente
    beneficio_relativo.sort(key=lambda x: (x[2], -x[0]))

    beneficio_total = 0 # Vamos a ir sumando el beneficio acumulado
    mochila = [] # Vamos a ir guardando los ítems que incluimos en la mochila

    for indice, beneficio, peso in beneficio_relativo:

        # Mientras la capacidad residual permita incluir el producto entero
        if capacidad >= peso:
            # Sumamos el beneficio del producto y restamos su peso de la capacidad
            beneficio_total += beneficio
            mochila.append((beneficio, peso))
            capacidad -= peso

        # Si el peso supera la capacidad disponible terminamos porque los ítems están ordenados por peso ascendente
        else:
            break

    return beneficio_total, mochila, capacidad
```

## Implementar una metaheurística Hill Climbing definiendo y justificando:


*   **Solución inicial**: partiremos de la primera heurística planteada en el punto anterior ya que obtuvo un mejor resultado, es decir, logró un beneficio más cercano al óptimo.
Sin embargo, teniendo en cuenta el gran rendimiento de esta heurística, también utilizaremos la segunda heurística para evaluar el desenvolvimiento del algoritmo a lo largo de las iteraciones.

*   **Función objetivo**: en este problema no hace falta plantear una función objetivo distinta a la original para poder definir qué tan buenos son los vecinos. La función objetivo que buscamos maximizar es la suma de los beneficios de los ítems incorporados a la mochila, siempre considerando la restricción de la capacidad de dicha mochila.
*   **Vecindad a recorrer**: se eliminará uno a uno cada elemento de la mochila inicial, reemplazándolo por ítems que originalmente habían quedado fuera, siempre priorizando los de mayor ratio beneficio/peso y respetando el límite de capacidad. Tendremos tantos vecinos como elementos tenga la mochila inicial.
*   **Criterio de terminación**:  Combinaremos dos criterios, iterando hasta alcanzar las mil iteraciones o hasta que la mejora sea menor o igual al 1%.

En conclusión, partiremos de una solución inicial, definiremos sus vecinos para evaluar si el mejor de ellos aporta una solución más beneficiosa y, en caso de serlo, volveremos a iterar buscando ahora los vecinos de esta nueva solución hasta alcanzar algún criterio de terminación.

Tengamos en cuenta que en el problema de la mochila completa, agregar artículos de alto ratio no nos asegura acercarnos a la solución óptima.

### B - Definimos la función para crear vencindarios.

```python
def vecindario (items_base, mochila_inicial, capacidad_max):

    # Definimos los items que quedaron fuera de nuestra mochila inicial
    fuera_mochila = [elemento for elemento in items_base if elemento not in mochila_inicial]

    # Ordenamos en forma descendente los items por ratio beneficio/peso
    fuera_mochila = [(item[0]/item[1], item[0], item[1]) for item in fuera_mochila]
    fuera_mochila.sort(reverse=True)
    fuera_mochila = [(item[1], item[2]) for item in fuera_mochila]

    vecinos = [] # Lista para almacenar los vecinos generados

    # Recorremos los elementos de la mochila inicial uno por uno
    for elemento_eliminar in mochila_inicial:

        # Generamos la mochila vecina con su capacidad
        mochila_vecina = mochila_inicial[:]
        capacidad_actual = sum(peso for _, peso in mochila_vecina)

        # Eliminamos el elemento de la mochila vecina y ajustamos capacidad
        mochila_vecina.remove(elemento_eliminar)
        capacidad_actual -= elemento_eliminar[1]

        # Recorremos los elementos de fuera_de_mochila para encontrar hasta alcanzar capacidad
        for elemento_agregar in fuera_mochila:
            if capacidad_actual + elemento_agregar[1] <= capacidad_max:
                mochila_vecina.append(elemento_agregar)
                capacidad_actual += elemento_agregar[1]
        vecinos.append(mochila_vecina)

    return vecinos
```

Evaluamos todos los vecinos generados por la función vencindario y nos quedamos con el mejor.


```python
# Función para calcular el beneficio total de una mochila dada
def calcular_beneficio(mochila):
    return sum(beneficio for beneficio, _ in mochila)
```

```python
def mejor_vecino(items_base, mochila_inicial, capacidad_max):

    # Generar los vecinos
    vecinos = vecindario(items_base, mochila_inicial,capacidad_max)

    # Calcular el beneficio de la mochila inicial
    beneficio_inicial = calcular_beneficio(mochila_inicial)

    # Inicializar la mejor mochila con la mochila inicial
    mejor_mochila = mochila_inicial
    mejor_beneficio = beneficio_inicial

    # Iterar a través de los vecinos
    for vecino in vecinos:
        # Calcular el beneficio del vecino
        beneficio_vecino = calcular_beneficio(vecino)

        # Comprobar si el vecino tiene un beneficio mejor que el mejor encontrado hasta ahora
        if beneficio_vecino > mejor_beneficio:
            mejor_mochila = vecino
            mejor_beneficio = beneficio_vecino

    return mejor_mochila
```

### C - Aplicamos el algoritmo en busca de maximizar beneficio hasta no encontrar una solución mejor.

Aplicamos el algoritmo de hill climbing calculando los vecinos de la mochila inicial, seleccionando el mejor de dichos vecinos e iterando hasta alcanzar las mil iteraciones o hasta que la mejora sea menor al 1%.

```python
def hill_climbing(items_base, mochila_inicial, capacidad_max, max_iter=1000, mejora_minima=0.01):

    mejores_mochilas = []  # Lista para almacenar las mejores mochilas de cada iteración
    mejores_mochilas.append((mochila_inicial, calcular_beneficio(mochila_inicial))) # Partimos de nuestra mochila inicial

    mejor_mochila = mochila_inicial
    mejor_beneficio = calcular_beneficio(mochila_inicial)

    for i in range(max_iter):
        # Generar vecinos
        vecinos = vecindario(items_base, mejor_mochila, capacidad_max)

        # Calcular el beneficio de la mochila actual
        beneficio_actual = calcular_beneficio(mejor_mochila)

        # Inicializar la mejor mochila en esta iteración
        mejor_vecino = None
        mejor_vecino_beneficio = mejor_beneficio

        # Iterar a través de los vecinos
        for vecino in vecinos:
            # Calcular el beneficio del vecino
            beneficio_vecino = calcular_beneficio(vecino)

            # Comprobar si el vecino tiene un beneficio mejor que el mejor encontrado hasta ahora
            if beneficio_vecino > mejor_vecino_beneficio:
                mejor_vecino = vecino
                mejor_vecino_beneficio = beneficio_vecino

        # Comprobar si se ha producido una mejora significativa
        if (mejor_vecino_beneficio - mejor_beneficio) / mejor_beneficio > mejora_minima:
            mejor_mochila = mejor_vecino
            mejor_beneficio = mejor_vecino_beneficio
            mejores_mochilas.append((mejor_mochila, mejor_beneficio))

        else:
            break  # Salir si no se ha producido una mejora significativa

    return mejores_mochilas
```

```python
# Ejemplo partiendo de una mochila lejos del óptimo y sin mejora mínima requerida

mochila_inicial = [(1,10), (1,10)]
mejores_mochilas = hill_climbing(items_10000, mochila_inicial, capacidad_10000, max_iter=1000, mejora_minima=0.0)

for i, (mochila, beneficio) in enumerate(mejores_mochilas):
    print("")
    print(f"Iteración {i}:")
    print("Mochila:", mochila)
    print("Beneficio:", beneficio)

print("")
print("Beneficio óptimo", base_10000_opt.iloc[0,0])
```

## Implementar un algoritmo genético para el problema definiendo y justificando:


*   **Población inicial**: Vamos a generar una población de cierta cantidad de mochilas creadas en forma random, respetando la restricción de la capacidad máxima de la mochila, pero buscando alcanzar alta versatilidad para no converger a soluciones parecidas rápidamente.
*   **Una función de fitness**: mantendremos la función objetivo, es decir, la
maximización del beneficio aportado por los items incluidos en la mochila. Será a partir de esta función que evaluaremos si una nueva solución es superadora.
*   **Un método de crossover**: Tomaremos un porcentaje de las mochilas con mayor beneficio y haremos crossover entre todas ellas, buscando generar nuevas mochilas a partir de las mejores. Realizaremos crossover de punto, es decir, definiremos en forma aleatoria un punto de corte (a nivel mochila es indistitnto donde cortemos), copiaremos todos los elementos antes del punto de corte del primer padre a la mochila descendiente y completaremos la capacidad restante con los elementos del segundo padre verificando que aún no formaban parte de la mochila descendiente. De esta manera, se garantiza que la mochila descendiente es factible y que contiene una combinación de elementos de ambos padres.
*   **Un método de mutación**: Aplicaremos mutaciones a un porcentaje de las mochilas con menor beneficio, buscando mejorar estos individuos y ganar en diversidad. Para esto, reemplazaremos un ítem random de la mochila por ítems que originalmente habían quedado fuera (elegidos también en forma aleatoria), siempre respetando el límite de capacidad.
*   **Un método de selección**: En el 10% inicial de iteraciones, la nueva generación será igual a su predecesora sumando los descendientes generados por crossover y las mutaciones. En estas iteraciones la cantidad de individuos será incremental para tener mayor variabilidad.
Luego, cada generación tendrá la misma cantidad de individuos que nuestra población inicial, seleccionando aquellos con mayor beneficio.
*   **Criterio de terminación**: Definiremos una cantidad de iteraciones a completar.



### Generamos una población inicial.

```python
def generar_mochilas_random(items_base, capacidad_max, num_mochilas=50):
    # random.seed(123) # Para mantener la reproductibilidad de los resultados
    mochilas_iniciales = []

    for _ in range(num_mochilas):
        mochila = []
        capacidad_actual = 0

        # Elegimos en forma random ítems y los agregamos a la mochila hasta alcanzar la capacida máxima
        elementos_random = random.sample(items_base, len(items_base)) # tomamos todos los items pero en un orden aleatorio
        for elemento in elementos_random:
            if capacidad_actual + elemento[1] <= capacidad_max:
                mochila.append(elemento)
                capacidad_actual += elemento[1]

        mochilas_iniciales.append(mochila)

    return mochilas_iniciales
```

```python
# Ejemplo
mochilas_iniciales = generar_mochilas_random(items_200, capacidad_200, num_mochilas=50)

# Imprimimos las primeras 10 mochilas generadas.
for i, mochila in enumerate(mochilas_iniciales[:10], start=1):
    print(f"Mochila {i}: {mochila}")
```

Nos quedaremos únicamente con el 10% de mejor fitness, es decir, de mayor beneficio.

```python
# Función para calcular el beneficio total de una mochila dada
def calcular_beneficio(mochila):
    return sum(beneficio for beneficio, _ in mochila)
```

```python
def seleccion_natural(mochilas_iniciales, capacidad_max, porcentaje=0.10):
    # Calculamos el beneficio de cada mochila y creamos una lista de tuplas (mochila, beneficio)
    mochilas_con_beneficio = [(mochila, calcular_beneficio(mochila)) for mochila in mochilas_iniciales]

    # Ordenamos las mochilas en función del beneficio en orden descendente
    mochilas_con_beneficio.sort(key=lambda x: x[1], reverse=True)

    # Calcula cuántas mochilas seleccionar (porcentaje del total)
    num_seleccionadas = int(len(mochilas_iniciales) * porcentaje)

    # Selecciona las mejores mochilas y nos quedamos únicamente con el elemento mochila de la tupla
    mejores_mochilas = [mochila for mochila, _ in mochilas_con_beneficio[:num_seleccionadas]]

    return mejores_mochilas
```

### Realizamos crossover entre los individuos seleccionados previamente.

```python
def crossover(padre1, padre2, capacidad_max):
    # Seleccionar un punto de corte aleatorio
    punto_de_corte = random.randint(1, len(padre1) - 1)

    # Crear una mochila descendiente inicialmente vacía
    descendiente = []

    # Copiar todos los elementos antes del punto de corte del padre 1
    for i in range(punto_de_corte):
        descendiente.append(padre1[i])

    # Completar la mochila descendiente con elementos del padre 2
    for elemento in padre2:
        if elemento not in descendiente and sum(peso for _, peso in descendiente) + elemento[1] <= capacidad_max:
            descendiente.append(elemento)

    return descendiente
```

```python
def crossover_entre_mejores(mejores_mochilas, capacidad_max):
    descendientes = []

    # Realizar el crossover entre todas las combinaciones posibles de mejores mochilas
    for i in range(len(mejores_mochilas)):
        for j in range(i + 1, len(mejores_mochilas)):
            descendiente = crossover(mejores_mochilas[i], mejores_mochilas[j], capacidad_max) # Hacemos el crossover
            descendientes.append(descendiente) # Guardamos el descendiente de cada cruce

    return descendientes
```

```python
# Ejemplo
nuevos_descendientes = crossover_entre_mejores(mejores_mochilas, capacidad_200)

# Imprime los descendientes resultantes
for i, descendiente in enumerate(nuevos_descendientes, start=1):
    print(f"Descendiente {i}: {descendiente}")
```

### Realizamos mutaciones en el 10% de las peores mochilas.

```python
def peores_individuos (mochilas_iniciales, capacidad_max, porcentaje=0.10):
    # Calculamos el beneficio de cada mochila y creamos una lista de tuplas (mochila, beneficio)
    mochilas_con_beneficio = [(mochila, calcular_beneficio(mochila)) for mochila in mochilas_iniciales]

    # Ordenamos las mochilas en función del beneficio en orden ascendente
    mochilas_con_beneficio.sort(key=lambda x: x[1], reverse=False)

    # Calcula cuántas mochilas seleccionar (porcentaje del total)
    num_seleccionadas = int(len(mochilas_iniciales) * porcentaje)

    # Selecciona las mejores mochilas y nos quedamos únicamente con el elemento mochila de la tupla
    peores_mochilas = [mochila for mochila, _ in mochilas_con_beneficio[:num_seleccionadas]]

    return peores_mochilas
```

```python
def mutacion(mochilas, items_base, capacidad_max):

    # Crear una lista para almacenar las mochilas mutadas
    mochilas_mutadas = []

    for mochila in mochilas:
        # Calcular los elementos disponibles excluyendo los elementos en la mochila
        elementos_disponibles = [elemento for elemento in items_base if elemento not in mochila]

        # Seleccionar un elemento aleatorio de la mochila para eliminar
        elemento_a_eliminar = random.choice(mochila)

        # Crear una copia de la mochila original para la mutación
        mochila_mutada = mochila.copy()

        # Eliminar el elemento seleccionado de la mochila mutada
        mochila_mutada.remove(elemento_a_eliminar)

        # Calcular la capacidad actual de la mochila
        capacidad_actual = sum(peso for _, peso in mochila_mutada)

        # Calcular la cantidad de espacio disponible para agregar nuevos elementos
        espacio_disponible = capacidad_max - capacidad_actual

        # Mientras haya espacio disponible y elementos disponibles para agregar
        while espacio_disponible > 0 and elementos_disponibles:
            # Seleccionar un elemento aleatorio de los disponibles que no esté en la mochila mutada
            elemento_nuevo = random.choice([elemento for elemento in elementos_disponibles])

            # Verificar si el peso del elemento nuevo no supera el espacio disponible
            if elemento_nuevo[1] <= espacio_disponible:
                # Agregar el elemento nuevo a la mochila mutada
                mochila_mutada.append(elemento_nuevo)
                espacio_disponible -= elemento_nuevo[1]
                # Eliminar el elemento nuevo de elementos_disponibles
                elementos_disponibles.remove(elemento_nuevo)
            else:
                # Si el elemento nuevo supera el espacio disponible, detener la mutación
                break

        # Agregar la mochila mutada a la lista de mochilas mutadas
        mochilas_mutadas.append(mochila_mutada)

    return mochilas_mutadas
```

```python
# Ejemplo
peores_mochilas = peores_individuos (mochilas_iniciales, capacidad_200, porcentaje=0.10)
mochilas_mutadas = mutacion(peores_mochilas, items_200, capacidad_200)

# Imprime las mutaciones resultantes
for i, mutaciones in enumerate(mochilas_mutadas, start=1):
    print(f"Mochila mutada {i}: {mutaciones}")
```

### Algoritmo genético.

```python
def algoritmo_genetico(items_base, capacidad_max, num_mochilas, max_iter, porcentaje):

    generaciones = []
    # tamaño = []
    generacion_actual = generar_mochilas_random(items_base, capacidad_max, num_mochilas)

    for iteracion in range(1, max_iter + 1):

        # Selección de las mejores mochilas
        mejores_mochilas = seleccion_natural(generacion_actual, capacidad_max, porcentaje)

        # Crossover entre las mejores mochilas
        descendientes_crossover = []
        for i in range(len(mejores_mochilas)):
            for j in range(i + 1, len(mejores_mochilas)):
                descendientes_crossover.append(crossover(mejores_mochilas[i], mejores_mochilas[j], capacidad_max))

        # Selección de las peores mochilas
        peores_mochilas = peores_individuos(generacion_actual, capacidad_max, porcentaje)

        # Mutación de las peores mochilas
        descendientes_mutacion = mutacion(peores_mochilas, items_base, capacidad_max)


        # Nueva generación
        if iteracion <= 0.1*max_iter:
            generacion_actual += descendientes_crossover + descendientes_mutacion
        else:
            generacion_actual += descendientes_crossover + descendientes_mutacion
            generacion_actual.sort(key=lambda x: -calcular_beneficio(x))
            generacion_actual = generacion_actual[:num_mochilas]

        generacion_actual.sort(key=lambda x: -calcular_beneficio(x))
        mejor_mochila_final = generacion_actual[:1]
        generaciones.append(mejor_mochila_final)
        # mochilas_totales = len(generacion_actual)
        # tamaño.append(mochilas_totales)

    return generaciones
```
